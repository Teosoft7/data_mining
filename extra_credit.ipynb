```python
alphabet_data = []

lowercase = ascii_lowercase 
uppercase = ascii_uppercase

for i in range(0, len(lowercase)):
    alpha_row = {'lower': lowercase[i],
                'upper': uppercase[i],
                'index': i}
    alphabet_data.append(alpha_row)
    
```

```python
alphabet_data
```

```python
import pandas as pd
from string import (ascii_lowercase, ascii_uppercase)

alphabet_df = pd.DataFrame(alphabet_data)
```
